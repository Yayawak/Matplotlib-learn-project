```python
import pandas as pd
import numpy as np
```

```python
uri = '../Sample data/SampleData.csv'
df = pd.read_csv(uri)
df.Total.replace(',', '', regex=True, inplace=True)
df.Total = df.Total.apply(pd.to_numeric)
df.head()
```

```python
df.dtypes
```

```python
pvt = df.pivot_table(index='Rep', columns='Item', fill_value=0)
pvt
```

```python
df.groupby('Item').sum()
```

```python
df.pivot_table(index=['Rep', 'Item'], margins=True)
```

```python
df.pivot_table(index=['Region'], values='Total')
```

```python
pvt2 = df.pivot_table(index='Rep', values='Total', columns='Item', fill_value=0)
pvt2
```

```python
pvt3 = df.pivot_table(index='Rep', 
                      values='Total', columns='Item', fill_value=0,
                     aggfunc=np.sum , margins=True)
# pvt3 = pvt.applymap('{0:,}'.format)
pvt3
```

```python
pvt = df.pivot_table(index='Rep', values='Total')
pvt.style.format({'Total': '{:,.2f}'})
pvt.applymap('${0:,.2f}'.format)
pvt
```

```python
pvt4 = df.pivot_table(index='Region', values='Total',
              aggfunc=['sum', 'max',
                      'min', 'mean'],
              margins=True)
pvt4
```

```python
import matplotlib.pyplot as plt
pvt4.plot.bar()
plt.show()
```

```python
pvt.plot.bar()
plt.show()
```
