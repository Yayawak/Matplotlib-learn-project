# Pivot Table

```python
df.dtypes
```

```python
# df.Total = df.Total.astype('float')
## bug : comma is not a number so can't convert str => float
```

```python
# df.Total.astype('string')
df.Total.replace(',', '', inplace=True, regex=True)
# convert str => num
# c = df.select_dtypes(object).columns
df['Total'] = df['Total'].apply(pd.to_numeric, errors='coerce')

df
```

```python
df.dtypes # successfully delete ',' from 'Total'
```

```python
# df = pd.read_csv(uri, dtype={'Total':"float64"}) 
# can't directly change when start read file because 
# it's still have ',' in original csv file
```

```python
pvt = df.pivot_table(index='Rep', margins=True, margins_name='Sum of total sales')
pvt
```

```python
pvt = df.pivot_table(index=['Rep', 'Item'], margins=True, margins_name='Sum of total sales')
pvt
```
