```python
missing_count = df.isnull().sum()
total_cells = np.product(df.shape)
total_cells
```

```python
total_missing = missing_count.sum()
total_missing
```

```python
missing_percent = (total_missing * 100) / total_cells
missing_percent # 2.5% = 5/200 --> 2.5 = (5 * 100) / 200
```

```python
print('missing: {} %'.format(missing_percent))
```

```python
print('missing: %.2f ' %missing_percent) # can't write literal %
```

```python
print(f'missing: {missing_percent.round(2)} %')
```
