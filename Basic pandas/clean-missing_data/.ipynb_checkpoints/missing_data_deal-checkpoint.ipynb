```python
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
import seaborn as sns

df = sns.load_dataset('titanic')
df.head(10)
# pclass = passenger class (1,2,3)
# sib / sp = sibling / spouse aboard
# par/ch = parent/ child abord
# fare = passenger fair
# embarked = port for embarkation c,q,s
# boat = life boat if survived
```

```python
df.dtypes
```

```python
df.isnull().any()
```

```python
missing_count = df.isnull().sum()
total_cells = np.product(df.shape)
total_cells
```

```python
total_missing = missing_count.sum()
total_missing
```

```python
missing_percent = (total_missing * 100) / total_cells
missing_percent # 2.5% = 5/200 --> 2.5 = (5 * 100) / 200
```

```python
print('missing: {} %'.format(missing_percent))
```

```python
print('missing: %.2f ' %missing_percent) # can't write literal %
```

```python
print(f'missing: {missing_percent.round(2)} %')
```

```python
col = ['age',]
df[col] = df[col].fillna(df[col].mean())

df.age.isnull().sum()
```

```python
to_drop = [ df.columns[i] for i in range(0,12)]
df_del = df.drop(columns=to_drop)
df_del
```
