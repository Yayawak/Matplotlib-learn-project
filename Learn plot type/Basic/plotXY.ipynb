```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-pastel')

freq = 8
amp = 2
# make data
x = np.linspace(0, 10, 100)
y = 4 + amp * np.sin(freq * x)

# plot
fig, ax = plt.subplots()

ax.plot(x, y, linewidth=2.0)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```
