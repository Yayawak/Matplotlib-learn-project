# Noise Reduction by rolling & resampling

```python
df.rolling(7, center=True).mean() # every 7 day -> do mean()
```

```python
df.rolling(7, center=True).sum()
```

```python
start, end = '2017-12-1','2017-12-7'
cols = ['Consumption', 'Wind', 'Solar']
df[cols][start:end].round(1)
```

```python
df_weekly_roll = df[cols].rolling(7, center=True).mean()
df_weekly_roll[start: end].round(1)
```

```python
df_weekly_resamp = df[cols].resample('7d').mean()
df_weekly_resamp
```

```python
df[cols].resample('3m').sum().plot()
```

```python
df[cols].resample('y').mean().plot()
```

```python
df_resamp = df[cols].resample('4w').mean()
df_resamp.plot(grid=True)
plt.ylabel('GWh')
plt.title('Power')
plt.show()
```

```python
start, end = '2017-2', '2017-10'
df_solar = df.Solar.loc[start: end]
df_solar.plot(marker='o', label='Daily', markersize=2, alpha=0.7)
df_solar.resample('m').mean().plot(marker='o', c='red',
                markersize=8, label='Mean Sampling')
df_solar.rolling(15, center=True).mean().plot(
        marker='o', markersize=2, label='Mean rolling'
        , c='g')

plt.ylabel('Power (GWh)')
plt.title('Solar')
plt.legend()
plt.grid()
plt.show()
```

```python
cols = ['Solar', 'Wind']
yearly = df[cols].resample('y').sum()

yearly.plot(kind='bar', figsize=(10,3), rot=40)
plt.grid()
plt.show()
```

```python
# delete excessive time 00:00:00 
def x_label(datetime):
    lb = str(datetime.year)
    return lb

ax = yearly.plot(kind='bar', figsize=(10,3), rot=20)
ax.set_xticklabels(map(lambda x : x_label(x), yearly.index ))
ax.set(xlabel='Year', ylabel='Power (GWh)')
ax.set(title='Yearly Power 2006-2018')
plt.grid()
```

```python
import numpy as np

range = pd.date_range('2020-01-1', '2020-2-28', freq='10min')
range
```

```python
df = pd.DataFrame(index=range)
df.head()
```

```python
df['temp'] = np.random.randint(low=10, high=40, size=df.shape[0])
df.head()
```

```python
df['humid'] = np.random.randint(50, 70, size=df.shape[0])
df.head()
```

```python
# df
# df.drop(columns='level_0', inplace=True)
# df
# df.reset_index(inplace=True)
df
```

```python
df.columns = ['time', 'temp', 'humid']
df.head()
```

```python
df.dtypes
```

```python
df.set_index('time', inplace=True)
df.head()
```

```python
df.loc['2020-01-15'].head()
```

```python
df.loc['2020-1-1 9:00': '2020-1-1 9:30']
```

```python
plt.rcParams['figure.figsize'] = 11,3

start = pd.to_datetime('2020-1-15 10:00')
end = pd.to_datetime('2020-1-18 1:00')

df.loc[start: end].plot(grid=True)
plt.xlabel('Date time')
plt.show()
```

```python
df_temp = df.temp.loc[start: end]
df_sampling = df.temp.loc[start: end].resample('1h').mean()
df_rolling = df.temp.loc[start: end ].rolling(6, center=True).mean()

df_temp.plot(grid=True, alpha=0.5)
df_sampling.plot(c='b', marker='o', label='Sampling Mean')
df_rolling.plot(c='r', label='Rolling Mean', grid=True)

plt.legend()
plt.xlabel('Date time')
plt.savefig('timeseries-datetime.png', dpi=120)
plt.show()
```
