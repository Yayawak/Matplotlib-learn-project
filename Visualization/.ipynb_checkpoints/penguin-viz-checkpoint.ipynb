```python
# want to calculate Sex m=1, f=-1, nan=0
import numpy as np
# nested ternary
df['sexnum'] = np.where(df.sex.str.match('Nan'), 0, 
                       (np.where(df.sex.str.match('Female'), -1, 1)))
df
```

```python
corr_df = df.corr().round(2)
corr_df
```

```python
df.groupby('sexnum').count()
```

```python
matrix = np.triu(corr_df)
sns.heatmap(corr_df, annot=True, mask=matrix, cmap='YlGnBu')
```

```python
matrix = np.triu(corr_df)
sns.heatmap(corr_df, annot=True, mask=matrix, )
```
