# Try all plotting Styles

```python
x = np.arange(1, 100, 10)
y = np.sin(x) 
fig = plt.figure(tight_layout=True, figsize=(10,20))
available = ['default'] + plt.style.available
for indx, style in enumerate(available):
    plt.style.use(style)
    with plt.style.context(style):
        ax = fig.add_subplot(10, 3, indx+1)
        ax.plot(x, y)
        ax.set_title(style)
```
