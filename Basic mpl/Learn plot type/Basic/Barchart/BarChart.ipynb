```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-deep')
```

```python
N = 5
# Score-grade of each group
menMeans = (20, 45, 30, 25, -35)
womenMeans = (18, 42, 35, 20, -33 )
menStd = ( 1, 4, 2, 1, 3)
womenStd = (4, 5, 2, 1, 2)
# index or location of each group
ind = np.arange(N)
width=0.35
```

```python
fig, ax = plt.subplots(figsize=[10,8])
pl1 = ax.bar(ind, menMeans, width, yerr=menStd, label='Men' )
pl2 = ax.bar(ind, womenMeans, width, yerr=womenStd,
            bottom=menMeans, label='Women')


ax.axhline(0, color='m', linewidth=3)
ax.set_ylabel('Scores')
ax.set_xlabel('Groups')
ax.set_title('Score by gender and group')
ax.set_xticks(ind, labels=['G1', 'G2', 'G3', 'G4', 'G5'])
ax.legend()

ax.bar_label(pl1, label_type='center')
ax.bar_label(pl2, label_type='center')

plt.show()
```
