```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 101)

fig = plt.figure(edgecolor=(0,0,0), linewidth=5)

ax1 = fig.add_subplot(235, title='$x^{1/2}$')
ax1.plot(x, np.sqrt(x), 'm')

ax2 = fig.add_subplot(232)
ax2.plot(x, np.log(x), 'g')
plt.show()
```

```python
x = np.arange(100)
fig, axes = plt.subplots(6, 3, figsize=[15,12])
for i in range(6):
    for k in range(3):
        axes[i,k].plot(x, np.sin(x*(i+k)), 'm')
plt.show()
```

# Better nested for loop

```python
x = np.arange(100)
fig, axes = plt.subplots(6, 2, figsize=[10,8], linewidth=25)
[ axes[i,j].set_ylim([-1,1]) for i in range(6) for j in range(2)]

for i in range(6):
    axes[i,0].plot(x, np.sin(x*(i+1)/2), '#BBAA33')
    axes[i,1].plot(x, np.cos(x*(i+1)/2), '#AABBAA')

plt.show()
```
