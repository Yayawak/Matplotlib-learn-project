```python
import matplotlib.pyplot as plt
import numpy as np

font = {
    'family': 'serif',
    'color': 'red',
    'size': 16,
    'weight': 'normal'
}

x = np.linspace(0, 10, 100)
y = np.sin(x *2 * np.pi)*np.exp(-x)

plt.plot(x, y, 'k')
tex = r'$sin(2\pi t)  e^{-t}$'
plt.text(7, 0.4, tex, fontdict=font)
plt.title('Damped exponential decay', fontdict=font)
plt.xlabel('time (s)', fontdict=font)
plt.ylabel('voltage (mV)', fontdict=font)

plt.show()
```
