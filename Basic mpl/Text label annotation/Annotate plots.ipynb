```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Ellipse
from matplotlib.text import OffsetFrom
```

```python
fig, ax = plt.subplots()
t = np.arange(0, 4, 0.01)
s = np.sin(2*np.pi*t)

line = ax.plot(t, s)
# only floating text, no arrowing
ax.annotate('figure pixels', xy=(10,10), xycoords='figure pixels', fontsize=15)


ax.annotate('point offset from data', xy=(2.25, 1)
           , xytext=(-15,25)
            ,xycoords='data',
           textcoords='offset points',
            horizontalalignment='right',
            verticalalignment='bottom',
           arrowprops=dict(facecolor='red', shrink=0.05),
            fontsize=20
           )
ax.annotate(r'$This\ is\ some\ of\ Min\grave{Y}$',
           xy=(2.75,-1), xytext=(.9,.5),
           textcoords='axes fraction', xycoords='data',
            fontsize=20,
           arrowprops=dict(facecolor='tab:green', shrink=0.1))
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
el = Ellipse((2, -1), 2, 1, facecolor='r')

# t = np.arange(0, 5, 0.1)
# # y = 1/16 * (np.sin(2*np.pi*t) * np.exp(t))
# y = np.sin(2*np.pi*t)
# ax.plot(t, y, linewidth=10)
ax.add_patch(el)
ax.set(xlim=(-1, 5), ylim=(-4, 2))

ax.annotate('$->$', xy=(1,-1), xytext=(0,-1), 
            bbox=dict(boxstyle='round',fc='0.8') ,
           arrowprops=dict(arrowstyle='->',
                          patchB=el,
                          connectionstyle='angle,angleA=90,angleB=0,rad=10'))
ax.annotate('arrow\nfancy', xy=(2,-.5), xytext=(-100,60),
            textcoords='offset points',
           size=20,
           arrowprops=dict(arrowstyle='fancy',
                          fc='0.6', ec='none',
                          patchB=el,
                          connectionstyle='angle3,angleA=0,angleB=-90'))
ax.annotate('arrow\nsimple', xy=(2.5,-.5), xytext=(0,70),
            textcoords='offset points',
           size=20,
           arrowprops=dict(arrowstyle='simple',
                          fc='0.6', ec='none',
                          patchB=el,
                          connectionstyle='arc3,rad=0.4'))
ax.annotate('wedge', xy=(2.5,-1.5), xytext=(100,-60),
            textcoords='offset points',
           size=20,
           arrowprops=dict(arrowstyle='wedge,tail_width=0.9',
                          fc='0.6', ec='none',
                          patchB=el,
                          connectionstyle='angle3,angleA=0,angleB=-90'))
an1 = ax.annotate('bubble', xy=(2,-1.5), xytext=(0,-80),
            textcoords='offset points',
           size=20,
            bbox=dict(boxstyle='round', fc='green'), va='center',
           arrowprops=dict(arrowstyle='wedge,tail_width=1',
                          fc='green', ec='none',
                          patchA=None,
                          patchB=el,
                          relpos=(0.2,0.5)))
an1.draggable()
ax.annotate('arrow\nfancy', xy=(2,-.5), xytext=(-100,60),
            textcoords='offset points',
           size=20,
           arrowprops=dict(arrowstyle='fancy',
                          fc='orange', ec='none',
                          patchB=el,
                          connectionstyle='angle3,angleA=0,angleB=-90'))
```
