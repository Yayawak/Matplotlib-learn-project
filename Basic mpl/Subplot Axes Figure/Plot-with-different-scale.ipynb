```python
import matplotlib.pyplot as plt
import numpy as np

# mock data 
t = np.arange(0, 10, 0.1)
f2 = np.sin(t * 2 * np.pi)
f1 = np.exp(t) # e^t

kwargs = dict(size=20)


fig, ax1 = plt.subplots(figsize=(8,6))

ax1.set_xlabel('time (s)', color='green', **kwargs)
ax1.tick_params(labelcolor='green', **kwargs)

ax1.plot(t, f1, color=c1)
c1 = 'tab:red'
ax1.set_ylabel('exp', color=c1, **kwargs)
ax1.tick_params(axis='y', labelcolor=c1, **kwargs)

c2 = 'orange'
ax2 = ax1.twinx()
ax2.plot(t, f2, color=c2)
ax2.set_ylabel('sine', color=c2, **kwargs)
ax2.tick_params(axis='y', labelcolor=c2, **kwargs)
plt.show()
```
