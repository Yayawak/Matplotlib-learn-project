```python
import matplotlib.pyplot as plt
import numpy as np

def example_plot(ax, fontsize=12, hide_labels=False):
    # pcolormesh simular to subplot but have color 
    pc = ax.pcolormesh(np.random.randn(30,30), vmin=-2.5, vmax=2.5)
    if not hide_labels: # if true
        ax.set_xlabel('x-label', fontsize=fontsize)
        ax.set_ylabel('y-label', fontsize=fontsize)
        ax.set_title('Title', fontsize=fontsize)
    return pc

fig = plt.figure(figsize=(14, 8))
subfigs = fig.subfigures(1, 2, wspace=0.07)
subfigs[0].set_facecolor('0.7')
subfigs[1].set_facecolor('0.6')

# left fig
axsLeft = subfigs[0].subplots(1, 2)
# return pcolormesh in 2 subplots
for ax in axsLeft: 
    pc = example_plot(ax)
subfigs[0].suptitle('Left fig', fontsize='xx-large')
subfigs[0].colorbar(pc, location='bottom', ax=axsLeft, shrink=0.8)

# right fig
axsRight = subfigs[1].subplots(3, 1)
subfigs[1].suptitle('Right fig', fontsize='xx-large')

for nn, ax in enumerate(axsRight):
    pc = example_plot(ax, hide_labels=True)
    if nn == 2:
        ax.set_xlabel('xlabel')
    if nn == 1:
        ax.set_ylabel('ylabel')
subfigs[1].colorbar(pc, ax=axsRight, shrink=0.8)
    
plt.show()
```
