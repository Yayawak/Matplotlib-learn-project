```python
import matplotlib.pyplot as plt
import numpy as np

# mock data
x = np.linspace(0, 2*np.pi, 360)
y = np.sin(x ** 1.5)

plt.plot(x,y)
```

```python
fig, axs = plt.subplots(1,2, figsize=(8,4), sharey=False)
plt.subplots_adjust(right=1.5)

fig.suptitle('Horizontal Stacked plots', size=20)
axs[0].plot(x,y,)

axs[1].set_ylim(-1,1)
axs[1].plot(x,np.abs(y))

axs2 = axs[0].twiny()
axs2.plot(x, np.abs(y), color='red')

plt.show()
```

## ShareXY 

```python
fig, (ax1, ax2) = plt.subplots(2, sharex=True)
ax1.plot(x, y)
ax2.plot(x-1, y)
```

## Remove Space Between Plots

```python
fig = plt.figure()
gs = fig.add_gridspec(3, hspace=0)
axes = gs.subplots(sharex=True, sharey=True)

fig.suptitle('Sharing both axes')

axes[0].plot(x, y, '-')
axes[1].plot(x, y**2, '+')
axes[2].plot(x, 0.3*y, '--')
```

### Basic Subplot at that time ( not like Subplot"s" which need to defined before )

```python
# mock data 2
x1 = np.linspace(0, 5)
y1 = np.cos( 2*np.pi*x1 ) * np.exp(-x1)
x2 = np.linspace(0, 2)
y2 = np.cos(x2*2*np.pi)

fig = plt.figure(figsize=(8,6))
fig.suptitle('2 Difference', fontsize=20)


plt.subplot(211)
plt.ylabel('Damped oscillation')
plt.plot(x1, y1, 'o-')

plt.subplot(212)
plt.ylabel('Undamped oscillation')
plt.plot(x2, y2, '.-')

plt.show()
```
