```python
import matplotlib.pyplot as plt
import numpy as np

fig1, ax = plt.subplots()
ax.set_xlim(2000, 3000)
ax.set_box_aspect(1) # square box
plt.show()
```

```python
fig, axes = plt.subplots(4,8, figsize=(10,10))
for i in range(4):
    for j in range(8):
        val = 1+i+j
        axes[i,j].set_box_aspect(val)
#         print(1+(i+j))
        axes[i,j].text(0.5, 0.5, val, fontsize=18)
```

## Normal plot next to image

```python
fig, (ax1,ax2) = plt.subplots(ncols=2, figsize=(20,18) ,
                             constrained_layout=True)

np.random.seed(19680801)  # Fixing random state for reproducibility
im = np.random.rand(9, 27)  # 2d array of rand 0 -> 1 # 3px* 27px
ax1.imshow(im)

ax2.plot([24, 80, 150], [3, 8, 20])
ax2.set_box_aspect(im.shape[0]/ im.shape[1]) # aspect = 3/27
plt.show()
```

```python
# Marginal plot (triple plot)
fig, axes = plt.subplots(2, 2, gridspec_kw=dict(height_ratios=[1,3],
                                               width_ratios=[3,1]),
                        figsize=(8,8),
                        sharex='col', # no need to write multiple ticks label
                        sharey='row'
                        )
axes[0,1].set_visible(False)
k = x,y = np.random.randn(2, 400) * [ [0.5], [180]] 
axes[1,0].scatter(x, y)
axes[0,0].hist(x)
axes[1,1].hist(y, orientation='horizontal')
plt.show()
```
