```python
import matplotlib.pyplot as plt

text_kwargs = dict(ha='center', va='center', fontsize=28, color='C3')
```

## Inch units ( defalut )

```python
plt.subplots(figsize=(6, 2)) # 6 inchs * 2 inchs
plt.text(0.5, 0.5 , '6 inchs * 2 inchs', **text_kwargs)
plt.show()
```

## Centimeter 

```python
cm = 1/2.54 # inch2cm
plt.subplots(figsize=(15,4))
plt.text(0.5, 0.5, '15 cm * 4 cm', **text_kwargs)
plt.show()
```

## Pixel Units
#### 1 px = 1/100 inch

```python
px = 1/plt.rcParams['figure.dpi']
plt.subplots(figsize=(800*px, 200*px))
plt.text(0.5, 0.5, '800px * 200px', **text_kwargs)
plt.show()
```
